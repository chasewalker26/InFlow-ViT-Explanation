```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('resid_ratio_10000_images.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

# residual 1 data
for i in range(len(data)):
    data[i] = np.asarray(data[i], dtype=float)

residual_1_ratio = data[0:12]
residual_2_ratio = data[12:24]

resid_1_layer_1 = np.array(data[0])
resid_1_layer_12 = np.array(data[11])

resid_2_layer_1 = np.array(data[12])
resid_2_layer_12 = np.array(data[23])

x = np.linspace(0, 197, 197)
plt.figure(figsize=(5, 3))
plt.rcParams.update({'font.size': 18})
plt.bar(x, resid_1_layer_1.squeeze(), color = "peru", align='center')
plt.xlim([-1, 198])
plt.title("Residual 1 Norm Ratios vs\nToken Number; Layer 1")
plt.xlabel("Token Number")
plt.ylabel("Input Residual : Attention")
plt.show()

x = np.linspace(0, 197, 197)
plt.figure(figsize=(5, 3))
plt.rcParams.update({'font.size': 18})
plt.bar(x, resid_1_layer_12.squeeze(), color = "peru", align='center')
plt.xlim([-1, 198])
plt.title("Residual 1 Norm Ratios vs\nToken Number; Layer 12")
plt.xlabel("Token Number")
plt.ylabel("Input Residual : Attention")
plt.show()

x = np.linspace(0, 197, 197)
plt.figure(figsize=(5, 3))
plt.rcParams.update({'font.size': 18})
plt.bar(x, resid_2_layer_1.squeeze(), color = "slateblue", align='center')
plt.xlim([-1, 198])
plt.title("Residual 2 Norm Ratios vs\nToken Number; Layer 1")
plt.xlabel("Token Number")
plt.ylabel("Attention Residual : MLP")
plt.show()

x = np.linspace(0, 197, 197)
plt.figure(figsize=(5, 3))
plt.rcParams.update({'font.size': 18})
plt.bar(x, resid_2_layer_12.squeeze(), color = "slateblue", align='center')
plt.xlim([-1, 198])
plt.title("Residual 2 Norm Ratios vs\nToken Number; Layer 12")
plt.xlabel("Token Number")
plt.ylabel("Attention Residual : MLP")
plt.show()
```
